```python
#total matches played in each season - graph
sns.countplot(match_data['Season'])
plt.xticks(rotation=90, fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel('Season',fontsize=10)
plt.ylabel('Count',fontsize=10)
plt.title('Total Matches Played in each season',fontsize=10,fontweight="bold")
```

```python
#No. of tosses won by each team
toss=match_data['TossWinner'].value_counts()
ax=plt.axes()
ax.set(facecolor="grey")
sns.set(rc={'figure.figsize':(15,10)},style='darkgrid')
ax.set_title('No. of tosses won by each team',fontsize=15,fontweight="bold")
sns.barplot(y=toss.index,x=toss,orient='h',palette="icefire",saturation=1)
plt.xlabel('# of tosses won')
plt.ylabel('Teams')
plt.show()    
```

```python
# most wins in IPL
temp = pd.DataFrame({"Winning_Team" : match_data['WinningTeam']})
count_wins = temp.value_counts()
labels = [X[0] for X in count_wins.keys()]
bar, ax = plt.subplots(figsize = (25, 20))
ax = plt.pie(x = count_wins, autopct = "%.1f%%", labels = labels)
plt.title("MOST WINS IN IPL", fontsize = 15)
plt.show()
```

```python
#Toss decision across seasons
ax=plt.axes()
ax.set(facecolor="grey")
sns.countplot(x='Season',hue='TossDecision',data=match_data,palette="magma",saturation=1)
plt.xticks(rotation=90,fontsize=10)
plt.yticks(fontsize=15)
plt.xlabel('\n Season',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.title('Toss decision across seasons',fontsize=12,fontweight="bold")
plt.show()
```

```python
#does choosing batting or balling first help in winning matches   
plt.figure(figsize=(12,4))
sns.countplot(match_data.TossDecision[match_data.TossWinner==match_data.WinningTeam])
plt.show()
```

```python
#toss decison of teams
teams = match_data['TossWinner'].unique()
decision_making = pd.DataFrame([], columns = ['Toss Winner', 'Decision', 'Times'])
for ID, element in enumerate(teams):
    temp_bat = match_data[(match_data['TossWinner'] == element) & (match_data['TossDecision'] == 'bat')]
    temp_field = match_data[(match_data['TossWinner'] == element) & (match_data['TossDecision'] == 'field')]
    decision_making = decision_making.append({'Toss Winner' : element, 'Decision' : 'bat', 'Times' : temp_bat['TossWinner'].count()}, ignore_index = True)  
    decision_making = decision_making.append({'Toss Winner' : element, 'Decision' : 'field', 'Times' : temp_field['TossWinner'].count()}, ignore_index = True) 
    
```

```python
decision_making
```

```python
#toss decision of teams - graph
sns.catplot(x = 'Toss Winner', y = 'Times', hue = 'Decision', data = decision_making, kind = 'bar', height = 5, aspect = 2)
plt.xticks(rotation = 90)
plt.title('TOSS DECISION OF TEAMS')
plt.xlabel('IPL TEAMS')
plt.ylabel('TOSS DECISION')
plt.show()
```
